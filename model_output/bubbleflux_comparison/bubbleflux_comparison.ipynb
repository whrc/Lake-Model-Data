```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
######## bubble flux on ###########
on_co2 = 'YKD-burned_bubble/co2_water  1  1f2.dat'
on_co2_bubble = 'YKD-burned_bubble/fbblflx_co2  1  1f2.dat'
on_ch4 = 'YKD-burned_bubble/methane_water  1  1f2.dat'
on_ch4_bubble = 'YKD-burned_bubble/fbblflx_ch4  1  1f2.dat'
on_ch4_soil = 'YKD-burned_bubble/methane_water_soil  1  1f2.dat'
on_doc = 'YKD-burned_bubble/DOC  1  1f2.dat'
```

```python
######## bubble flux off ###########
off_co2 = 'YKD-burned_nobubble/co2_water  1  1f2.dat'
off_co2_bubble = 'YKD-burned_nobubble/fbblflx_co2  1  1f2.dat'
off_ch4 = 'YKD-burned_nobubble/methane_water  1  1f2.dat'
off_ch4_bubble = 'YKD-burned_nobubble/fbblflx_ch4  1  1f2.dat'
off_ch4_soil = 'YKD-burned_nobubble/methane_water_soil  1  1f2.dat'
off_doc = 'YKD-burned_nobubble/DOC  1  1f2.dat'
```

```python
df_on_ch4 = pd.read_csv(on_ch4, delimiter=r"\s+", skiprows=7, index_col=None, header=None)
df_on_ch4.columns=['year', 'month', 'day', 'hour', 'integration_time', 'depth', 'ch4']
df_on_ch4['datetime'] = pd.to_datetime({'Year': df_on_ch4['year'], 'Month': df_on_ch4['month'], 'Day':df_on_ch4['day']})
df_on_ch4 = df_on_ch4[df_on_ch4['ch4']!=-999]

df_off_ch4 = pd.read_csv(off_ch4, delimiter=r"\s+", skiprows=7, index_col=None, header=None)
df_off_ch4.columns=['year', 'month', 'day', 'hour', 'integration_time', 'depth', 'ch4']
df_off_ch4['datetime'] = pd.to_datetime({'Year': df_off_ch4['year'], 'Month': df_off_ch4['month'], 'Day':df_off_ch4['day']})
df_off_ch4 = df_off_ch4[df_off_ch4['ch4']!=-999]
```

```python
sns.lineplot(data=df_on_ch4, x='datetime', y='ch4', label='bubble on')
sns.lineplot(data=df_off_ch4, x='datetime', y='ch4', label='bubble off')
```
