```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path_to_burned_soil_2022_23 = '15B_FP_ST-SM_Profile_22-23.xlsx'
path_to_unburned_soil_2022_23 = 'UB_FP_ST-SM_Profile_22-23.xlsx'

path_to_burned_soil_24 = '15B_FP_ST-SM_Profile_24.xlsx'
path_to_unburned_soil_24 = 'UB_FP_ST-SM_Profile_24.xlsx'

burned_soil_2022_23 = pd.read_excel(path_to_burned_soil_2022_23, parse_dates=['Date Time, GMT-09:00']).rename(columns={'Date Time, GMT-09:00': 'Date'})
unburned_soil_2022_23 = pd.read_excel(path_to_unburned_soil_2022_23, parse_dates=['Date Time, GMT-09:00']).rename(columns={'Date Time, GMT-09:00': 'Date'})

burned_soil_24 = pd.read_excel(path_to_burned_soil_24, parse_dates=['Date Time, GMT-09:00']).rename(columns={'Date Time, GMT-09:00': 'Date'})
unburned_soil_24 = pd.read_excel(path_to_unburned_soil_24, parse_dates=['Date Time, GMT-09:00']).rename(columns={'Date Time, GMT-09:00': 'Date'})

burned_soil_2022_23.head()
```

```python
burned_soil_2022_23_daily_mean = burned_soil_2022_23.groupby(by=burned_soil_2022_23['Date'].dt.date).mean()
burned_soil_2022_23_daily_std = burned_soil_2022_23.groupby(by=burned_soil_2022_23['Date'].dt.date).std()
burned_soil_2024_daily_mean = burned_soil_24.groupby(by=burned_soil_24['Date'].dt.date).mean()
burned_soil_2024_daily_std = burned_soil_24.groupby(by=burned_soil_24['Date'].dt.date).std()

burned_soil_2022_23_daily = pd.merge(burned_soil_2022_23_daily_mean, burned_soil_2022_23_daily_std, left_index=True, right_index=True, suffixes=('_mean', '_std')).reset_index()
burned_soil_2022_23_daily['pond'] = 'burned'

burned_soil_2024_daily = pd.merge(burned_soil_2024_daily_mean, burned_soil_2024_daily_std, left_index=True, right_index=True, suffixes=('_mean', '_std')).reset_index()
burned_soil_2024_daily['pond'] = 'burned'


unburned_soil_2022_23_daily_mean = unburned_soil_2022_23.groupby(by=unburned_soil_2022_23['Date'].dt.date).mean()
unburned_soil_2022_23_daily_std = unburned_soil_2022_23.groupby(by=unburned_soil_2022_23['Date'].dt.date).std()

unburned_soil_2024_daily_mean = unburned_soil_24.groupby(by=unburned_soil_24['Date'].dt.date).mean()
unburned_soil_2024_daily_std = unburned_soil_24.groupby(by=unburned_soil_24['Date'].dt.date).std()

unburned_soil_2022_23_daily = pd.merge(unburned_soil_2022_23_daily_mean, unburned_soil_2022_23_daily_std, left_index=True, right_index=True, suffixes=('_mean', '_std')).reset_index()
unburned_soil_2022_23_daily['pond'] = 'unburned'

unburned_soil_2024_daily = pd.merge(unburned_soil_2024_daily_mean, unburned_soil_2024_daily_std, left_index=True, right_index=True, suffixes=('_mean', '_std')).reset_index()
unburned_soil_2024_daily['pond'] = 'unburned'

soil_data = pd.concat([burned_soil_2022_23_daily, unburned_soil_2022_23_daily, burned_soil_2024_daily, unburned_soil_2024_daily])
```

```python
soil_data.to_csv('YKD_soil_data_2022_24.csv', index=False)
```
